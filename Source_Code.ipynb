# Chi-Square Test of Independence

```
# Checking that is there any relationship between Data Source and Description of Cargo?
# Null Hypothesis(Ho) : There is no relationship between Data Source and Description of Cargo
# Alternate Hypothesis(H1): There is a significant relationship between Data Source and Description of Cargo
from scipy.stats import chi2_contingency
# 1. Observation Table:
o_table = pd.crosstab(dataframe_copy["Data Source"],dataframe_copy["Description of Cargo"])
# 2. Expected values,chi_square value, Degrees of Freedom, p-values
chi2,p,df,expected = chi2_contingency(o_table)
# Values to watch
print(f"Chi_Square Value: {chi2}")
print(f"p Value: {p}")
# Decision
if p<0.05:
  print("Ho rejected. And there is a significant relationship between Data Source and Description of Cargo")
else:
  print("Ho accepted. And there is no significant relationship between Data Source and Description of Cargo")
```

```
# Checking that is there any relationship between Data Source and the Crossing?
# Null Hypothesis(Ho): There is no relationship between Data Source and the Crossing
# Alternate Hypothesis(H1): There is a significant relationship between Data Source and the Crossing
from scipy.stats import chi2_contingency
# 1. Observation Table:
o_table = pd.crosstab(dataframe_copy["Data Source"],dataframe_copy["Crossing"])
# 2. Expected value, chi_square value, Degrees of Freedom, p-values
chi2,p,df,expected = chi2_contingency(o_table)
# Vlues to watch
print(f"Chi square value: {chi2}")
print(f"p value: {p}")
# Decision
if p<0.05:
  print("Ho rejected. And there is a significant relationship between Data Source and the Crossing")
else:
  print("Ho accepted. And there is no significant relationship between Data Source and the Crossing")
```

```
# Checking that is there any relationship between Description of Cargo and the Crossing?
# Null Hypothesis(Ho) : There is no relationship between Description of Cargo and the Crossing
# Alternate Hypothesis(H1): There is a significant relationship between Description of Cargo and the Crossing
from scipy.stats import chi2_contingency
# 1. Observation Table:
o_table = pd.crosstab(dataframe_copy["Description of Cargo"],dataframe_copy["Crossing"])
# 2. Expected values,chi_square value, Degrees of Freedom, p-values
chi2,p,df,expected = chi2_contingency(o_table)
# Values to watch
print(f"Chi_Square Value: {chi2}")
print(f"p Value: {p}")
# Decision
if p<0.05:
  print("Ho rejected. And there is a significant relationship between Description of Cargo and the Crossing")
else:
  print("Ho accepted. And there is no significant relationship between Description of Cargo and the Crossing")
```
